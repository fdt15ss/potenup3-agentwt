```python
from typing import Literal
from pydantic import BaseModel, Field
from langchain.chat_models import init_chat_model


class CineReview(BaseModel):
    rating: int | None = Field(description="평점", ge=1, le=5)
    sentiment: Literal["positive", "negative"] = Field(description="감성")
    key_point: list[str] = Field(description="핵심포인트, 소문자, 1~3단어")
    title: str = Field(description="영화 제목")


model = init_chat_model("gpt-5.4-mini")
structured_model = model.with_structured_output(schema=CineReview)
review = """오늘 보았던 영화 귀멸의 칼날은 일본에서 역대 성적 1위를 기록하고 있는 센과 치히로의 행방불명을 뛰어넘은 흥행 1위 귀멸의 칼날 극장판 무한 열차입니다. 기존에 만화책으로도 먼저 보기도 했고 중간에 넷플릭스로 최근 보면서 기존에 내용을 어느 정도 숙지하고 있는 상태라 극장판을 보면서 이해하는 데는 문제가 없었습니다. 하지만 이 만화를 모르는 시점에서 보는 분이라면 네츠코가 왜? 저 아이는 왜 저런 기술을? 왜 싸움? 이런 생각을 들 수 있기 때문에 좀 더 몰입감을 높이기 위해서는 꼭 사전 내용을 보고 가거나 줄거리를 알고 가시는 것을 추천드립니다.
실제 접한 극장판의 퀄리티는 기존의 애니메이션보다 훨씬 퀄리티가 높았습니다. 아무래도 귀멸 칼날의 액션씬은 최근 나온 애니메이션 중에서 손에 꼽힌다고 느낄 수 있을 정도였습니다. 탄 지로가 주인공이긴 하나 다른 주 렌코 쿠의 일대기와 그의 힘을 극장판으로 무한의 액션감을 느낌수 있었네요 ㅎㅎ 최근에 여러 영화를 봤지만.... 애니메이션이 더 재미있었던 건 왜일까요 ㅎㅎ 기존의 애니메이션에 거부감이 없고 액션을 즐기시는 분이라면 재미있게 보실 수 있으니 추천드리고 싶습니다 ^^"""
result = structured_model.invoke(review)
result
```

```python
text = """아이사 클라우드싱어

투슈이의 길을 걷기로 한 우리가 얼라이언스에 충성을 맹세한 게 마치 어제와 같습니다.

안두인 국왕을 전적으로 존중하기는 하나, 솔직히 말해 그의 아버지가 훨씬 더 강력한 동맹이 되어 주지 않았나 싶습니다.

안두인 국왕에게는 말하지 말아 주십시오.

<몸을 숙이는 아이사>"""


class CheckTheText(BaseModel):
    npc_name: str = Field(description="대사하는 캐릭터")
    emotion: str = Field(description="감정")
    quest_related: list[str] = Field(description="관련 퀘스트 2~3개")
    lore_tags: str = Field(description="게임 제목")


model = init_chat_model("gpt-5.4-mini")
structured_model = model.with_structured_output(schema=CheckTheText)

result = structured_model.invoke(text)
result
```
