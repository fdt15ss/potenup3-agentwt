```python
import os, urllib.request, urllib.parse, json
from dotenv import load_dotenv

load_dotenv()

CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")


def naver_api(url: str):
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", CLIENT_ID)
    request.add_header("X-Naver-Client-Secret", CLIENT_SECRET)

    response = urllib.request.urlopen(request)
    return json.loads(response.read().decode("utf-8"))
```

```python
def get_news_summary(query: str):
    enc = urllib.parse.quote(query)
    url = (
        f"https://openapi.naver.com/v1/search/news.json?query={enc}&display=5&sort=date"
    )

    data = naver_api(url)

    summaries = []
    for item in data["items"]:
        title = item["title"]
        desc = item["description"]
        link = item["link"]

        summaries.append(
            f"""
            제목: {title}
            요약: {desc}
            링크: {link}
            """
        )

    return "\n".join(summaries)
```

```python
def get_book_info(query: str):
    enc = urllib.parse.quote(query)
    url = f"https://openapi.naver.com/v1/search/book.json?query={enc}&display=5"

    data = naver_api(url)

    results = []
    for item in data["items"]:
        title = item["title"]
        author = item["author"]
        price = item["discount"]
        link = item["link"]

        results.append(
            f"""
제목: {title}
저자: {author}
가격: {price}원
구매링크: {link}
"""
        )

    return "\n".join(results)
```

```python
def agent(user_input: str):
    if "뉴스" in user_input:
        return get_news_summary(user_input.replace("뉴스", "").strip())

    elif "책" in user_input:
        return get_book_info(user_input.replace("책", "").strip())

    else:
        return "뉴스 또는 책을 포함해서 질문해주세요."
```

```python
print(agent("AI 뉴스"))
print(agent("해리포터 책"))
```
