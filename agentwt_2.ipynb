```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import SystemMessage, HumanMessage

model = init_chat_model("gpt-5.4-mini")

msg = [
    SystemMessage("당신은 인공지능 에이전트입니다. 모르는 건 답하지 말아주세요"),
    HumanMessage("머신러닝이란?"),
]
model.invoke(msg)
```

```python
from langchain.agents import create_agent
from langchain.tools import tool

@tool
def search_book(title:str) -> str:
    '''책 제목을 입력받아 검색 결과를 반환'''
    books = {
        '파이썬': '작가: 김한호, 가격 : 100000',
        'AI': '작가 : 김한호, 가격 : 600000',
    }
    return books.get(title, '검색 결과가 없습니다.')

@tool
def get_weather(city: str) -> str:
    ''' 도시 이름을 입력받아 해당 도시의 날씨 정보를 반환 '''
    weather_data = {
        ['서울'] : '맑음',
        ['부산'] : '흐림'
    }
    return weather_data.get(city, '해당 도시의 정보가 없습니다.')

@tool
def add_number(a: int, b: int) -> int:
    """두 숫자를 더해서 결과를 전달"""
    return a + b


agent = create_agent(
    model="gpt-5.4-mini",
    tools=[get_weather, add_number],
    system_prompt="You are a helpful assistant",
)

result = agent.invoke({"messages": [{"role": "user", "content": "3+5+8은?"}]})
```

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun, YouTubeSearchTool

search = DuckDuckGoSearchRun()
youtube = YouTubeSearchTool()


agent = create_agent(
    model="gpt-5.4-mini",
    tools=[search, youtube],
    system_prompt="You are a helpful assistant",
)

result = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": "언더테일 노래 찾아서 유튜브 주소로 알려줘"}
        ]
    }
)
```
